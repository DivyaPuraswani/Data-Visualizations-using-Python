Pie charts are good to show proportional data of different categories and figures/numbers are usually in percentages. here, Distribution of cookies, jellybeans etc is represented. 

```python
import matplotlib.pyplot as plt

labels = ['Cookies', 'Jellybean', 'Milkshake', 'Cheesecake']
sizes = [38.4, 40.6, 20.7, 10.3]
colors = ['#235175', '#3A88C4', 'lightskyblue', '#D7E7F3']   #Using color scheme Blues
fig = plt.figure(figsize=(10,5),dpi=100)
ax= fig.gca()
patches, texts = plt.pie(sizes, colors=colors, shadow=True, startangle=90)
plt.legend(patches, labels, loc="best")
plt.axis('equal')
plt.tight_layout()
plt.title('Distribution of different varieties of sweets')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.datasets import load_breast_cancer
sns.set()
cancer = load_breast_cancer()
breast_cancer_df = pd.DataFrame(cancer['data'])
breast_cancer_df.columns = cancer['feature_names']
breast_cancer_df['target'] = cancer['target']
breast_cancer_df['diagnosis'] = [cancer['target_names'][x] for x in cancer['target']]
sns.set()
corr = breast_cancer_df[list(cancer['feature_names'])].corr(method='pearson')

cols = ['worst concave points', 'mean concavity', 
        'worst perimeter', 'worst radius', 
        'worst area']

sns.pairplot(breast_cancer_df,
             x_vars = cols,
             y_vars = cols,
             hue = 'diagnosis', 
             palette = ('Blue', '#875FDB'), 
             markers=["o", "D"])
```
