Violin Plot is used to visualize the distribution of numerical data of different variables. It is similar to Box Plot but with a rotated plot on each side, giving more information about the density estimate on the y-axis.
The density is mirrored and flipped over and the resulting shape is filled in, creating an image resembling a violin. The advantage of a violin plot is that it can show nuances in the distribution that arenâ€™t perceptible in a boxplot. On the other hand, the boxplot more clearly shows the outliers in the data.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
df = sns.load_dataset('tips')
print(df.head(20))
```

Considering tips dataset. Here days are plotted on x axis while total_bill on y axis. Mentioning hue as smoker , and data has smoker only 2 values as yes and no, which will plot the distibution of total_bills for each days for two distinct values of smoker [Yes,No]. Palette in seaborn represent the colour palette using Blues.

```python
fig = plt.figure(figsize=(10,5),dpi=100)   #setting figure size which is considered as one of the pre attentive property
ax= fig.gca()
ax.set_title('Total Bill for each day for Smoker') #setting title for the figure
plt.grid(False)   #Gridlines-False
sns.violinplot(x="day", y="total_bill", hue="smoker", data=df, palette="Blues")
```
