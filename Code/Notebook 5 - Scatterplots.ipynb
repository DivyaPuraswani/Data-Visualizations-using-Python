Relationship charts are used to show a possible connection or correlation between two or more variables. Scatter plot can be plotted by matplotlib or seaborn. Using matplotlib, .plot.scatter() method in pandas, or plt.scatter(). Using seaborn, sns.scatter() function can be called. As an example, the plot below plots the defense vs attach values for 1st-generation Pokemons.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pokemon=pd.read_csv('pokemon.csv')
```

```python
pokemon.head()
```

```python
g1 = pokemon.loc[pokemon.generation==1,:]
```

```python
g1.head()
```

```python
g1.plot.scatter('attack', 'defense', s = 40, c = 'blue', marker = 's', figsize=(8,5.5), title="Defence vs Attack for 1st Generation pokemons");
```

Using sns

```python
plt.figure(figsize=(10,7))
plt.title('Defence vs Attack for 1st Generation pokemons')
sns.scatterplot(x = 'attack', y = 'defense', s = 70, hue ='is_legendary', data=pokemon);
```
