Pick the right chart: Comparison charts are used to compare two variables, categories, or different items.In this notebook, we have a standard bar chart/graph and its other variations like double bar chart, stacked bar chart and horizontal bar chart using the Python library Matplotlib. 

Standard bar graph: This plot represents the acceleration due to gravity $g$ on Mercury (3.76$m/s^2$), Venus (9.04$m/s^2$), Earth (9.8$m/s^2$) and Mars (3.77$m/s^2$).

```python
import matplotlib.pyplot as plt
import numpy as np
x = ['Mercury', 'Venus', 'Earth', 'Mars']
g = [3.76,9.04,9.8,3.77]
x_pos = np.arange(len(x))
plt.figure(figsize=(10,10))
ax = plt.gca()
plt.bar(x_pos, g, color='#7ed6df')
plt.xlabel("Planets")
plt.ylabel("g")
plt.title("Acceleration Due to Gravity")
plt.xticks(x_pos, x)
plt.grid(False)
ax.set_facecolor('xkcd:white')
plt.show()
```

Horizontal bar chart:This plot represent the maximum speeds of the following hyper cars as a horizontal bar graph: Bugatti Chiron (420 km/h), Hennessey Venom F5 (435 km/h) and Koenigsegg Agera RS (457 km/h).  The left margin for the $y$-axis is achieved by setting left=0.3 inside the subplots_adjust() function.

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
cars = ('Bugatti Chiron', 'Hennessey Venom F5', 'Koenigsegg Agera RS')
y = np.arange(len(cars))
speeds = [420,435,457]
ax.barh(y, speeds, align='center',
        color='#7ed6df', ecolor='black')
ax.set_yticks(y)
ax.set_yticklabels(cars)
ax.set_facecolor('xkcd:white')
ax.invert_yaxis()
ax.set_xlabel('Top Speed (km/h)')
plt.subplots_adjust(left=0.3)
plt.title("Top Speed of cars")
plt.show()
```

Stacked bar charts: This plot displays the final results of the men's doubles badminton tournament at the 2016 Summer Olympics at Riocentro, Brazil, between the shuttlers from China and Malaysia scoring 16-21, 21-11, 23-21 in the first, second and third rounds respectively.

```python
import numpy as np
import matplotlib.pyplot as plt
n = 3
china = (16,21,23)
malaysia = (21,11,21)
ax = plt.gca()
x = np.arange(n)
width = 0.35
#ax = df_top20.plot.barh(stacked=True, figsize=(15, 12), cmap=plt.cm.get_cmap('tab20c'))
p1 = plt.bar(x, malaysia, width, color='cornflowerblue')
p2 = plt.bar(x, china, width, color='#7ed6df', bottom=malaysia)
plt.ylabel('Points')
plt.title('China v/s Malaysia')
plt.xticks(x, ('Round 1', 'Round 2', 'Round 3'))
plt.yticks(np.arange(0, 60, 5))
plt.legend((p2[0], p1[0]), ('China', 'Malaysia'))
ax.set_facecolor('xkcd:white')
plt.show()
```
