A heat map (or heatmap) is a graphical representation of data which can be created using Matplotlib and numpy where the individual values contained in a matrix are represented as colors.The goal of the heatmap is to provide a colored visual summary of information.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
flights = sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")
fig = plt.figure(figsize=(10,5),dpi=100)
ax= fig.gca()
ax = sns.heatmap(flights,annot=True, fmt="d",cmap='Blues',linewidths=.5)
ax.set_title('Count of passengers monthtly and yearly')
```
