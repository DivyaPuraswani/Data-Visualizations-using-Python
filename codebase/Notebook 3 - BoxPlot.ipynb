A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile (Q1), median, third quartile (Q3), and “maximum”). It can tell about your outliers and what their values are. It can also tell if your data is symmetrical, how tightly your data is grouped, and if and how your data is skewed.

```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
```

```python
movies= pd.read_csv("imdb_1000.csv")
```

```python
movies.head()
```

```python
movies.duration.plot(kind='box');
```

```python
import seaborn as sns
```

```python
df = pd.read_csv("mpg.csv")

plt.figure(figsize=(13,10), dpi= 80)
sns.boxplot(x='class', y='hwy', data=df, notch=False)
plt.title('Box Plot showing Highway Mileage by Vehicle Class', fontsize=22)
plt.ylim(10, 40)
plt.show()
```
